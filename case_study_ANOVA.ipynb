# 3.Check whether all zones generated same amount of sales!?

Null Hypothesis(H0)=All zones generate same amount of sales.

Alternative Hypothesis(H1)=Not all zones generate same amount of sales.

significance level=0.05

```python
data.head()
```

```python
from scipy import stats
```

```python
F,p=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print('F-test:',round(F,4),'p-value:',round(p,4))
```

```python
if p >= 0.05:
    print("Accept null hypothesis")
else:
    print("Reject null hypothesis")
```

INSIGHT:
    
One-way ANOVA test is performed.
    
Therefore (P-value < 0.05 ).So we reject null hypothesis

Not all companies generate same amount of sales
